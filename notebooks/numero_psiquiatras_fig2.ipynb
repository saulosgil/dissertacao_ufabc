```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Dados: (país, valor)
dados = [
    ("Suíça", 53.00),
    ("Alemanha", 28.00),
    ("Grécia", 26.00),
    ("Lituânia", 25.00),
    ("Países Baixos", 25.00),
    ("Noruega", 24.00),
    ("França", 23.00),
    ("Irlanda", 23.00),
    ("Áustria", 22.00),
    ("Nova Zelândia", 22.00),
    ("Suécia", 22.00),
    ("Luxemburgo", 21.00),
    ("Estônia", 20.00),
    ("Islândia", 20.00),
    ("Itália", 20.00),
    ("Reino Unido", 20.00),
    ("Austrália", 19.00),
    ("Croácia", 19.00),
    ("Dinamarca", 19.00),
    ("Canadá", 18.00),
    ("Finlândia", 18.00),
    ("MÉDIA PAÍSES", 17.83),
    ("Bélgica", 17.00),
    ("Eslovênia", 17.00),
    ("Israel", 17.00),
    ("República Tcheca", 17.00),
    ("Romênia", 16.00),
    ("Estados Unidos", 15.00),
    ("Hungria", 15.00),
    ("Letônia", 15.00),
    ("Portugal", 15.00),
    ("Chipre", 14.00),
    ("Chile", 13.00),
    ("Espanha", 13.00),
    ("Japão", 13.00),
    ("Polônia", 13.00),
    ("Bulgária", 10.00),
    ("Coreia do Sul", 9.00),
    ("Turquia", 7.00),
    ("BRASIL", 6.69),
    ("Colômbia", 2.00),
    ("México", 1.00),
]

paises = [d[0] for d in dados]
valores = [d[1] for d in dados]

# Cores por categoria
cores = []
for p, v in dados:
    if p == "BRASIL":
        cores.append("Blue")        # Verde destaque
    elif p == "MÉDIA PAÍSES":
        cores.append("red")        # Vermelho para a média
    else:
        cores.append("#5BB8B0")        # Verde-azulado padrão

fig, ax = plt.subplots(figsize=(13, 16))
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

# Barras horizontais (ordem de cima para baixo = índice invertido)
y_pos = np.arange(len(paises))
bars = ax.barh(y_pos, valores, color=cores, height=0.65, zorder=3)

# Eixo Y: nomes dos países
ax.set_yticks(y_pos)
ax.set_yticklabels(paises, fontsize=9.5)

# Colorir rótulos especiais
for i, (tick, pais) in enumerate(zip(ax.get_yticklabels(), paises)):
    if pais == "BRASIL":
        tick.set_color("Blue")
        tick.set_fontweight("bold")
        tick.set_fontsize(10.5)
    elif pais == "MÉDIA PAÍSES":
        tick.set_color("red")
        tick.set_fontweight("bold")
        tick.set_fontsize(10)

# Inverter eixo para maior valor no topo
ax.invert_yaxis()

# Valores ao lado das barras
for i, (v, p) in enumerate(zip(valores, paises)):
    cor_texto = "Blue" if p == "BRASIL" else ("red" if p == "MÉDIA PAÍSES" else "#333333")
    peso = "bold" if p in ("BRASIL", "MÉDIA PAÍSES") else "normal"
    ax.text(v + 0.4, i, f"{v:,.2f}".replace(".", ","),
            va="center", ha="left", fontsize=9, color=cor_texto, fontweight=peso)

# Grade vertical suave
ax.xaxis.grid(True, linestyle="--", alpha=0.4, color="gray", zorder=0)
ax.set_axisbelow(True)

# Limites do eixo X
ax.set_xlim(0, 62)
ax.set_xlabel("Médicos Especialistas em Psiquiatria por 100.000 habitantes", fontsize=10, color="#555555", labelpad=10)

# Linha de referência da média
media_idx = paises.index("MÉDIA PAÍSES")
ax.axvline(x=17.83, color="red", linewidth=1.2, linestyle="--", alpha=0.6, zorder=2)

# Destaque visual para a barra do Brasil
brasil_idx = paises.index("BRASIL")
ax.get_yticklabels()[brasil_idx].set_fontweight("bold")

# Seta apontando para o Brasil
ax.annotate(
    "Brasil:\nAbaixo da\nmédia global",
    xy=(9, brasil_idx),
    xytext=(25, brasil_idx - 1.5),
    fontsize=12,
    color="Blue",
    fontweight="bold",
    arrowprops=dict(arrowstyle="->", color="Blue", lw=1.4),
    ha="center",
    va="center",
)

# Remover bordas desnecessárias
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(True)
ax.spines["bottom"].set_visible(True)

# Título
fig.suptitle(
    "Médicos especialistas em Psiquiatria por 100.000 habitantes,\nsegundo países selecionados, em 2023",
    fontsize=13,
    fontweight="bold",
    color="#222222",
    x=0.5,
    y=0.97,
    ha="center",
    va="top",
)
ax.set_title("Fonte: Scheffer, 2025", fontsize=8.5, color="#888888", pad=6)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.savefig("../img/psiquiatras_por_pais.png", dpi=300, bbox_inches="tight", facecolor="white")
plt.show()
print("Gráfico salvo com sucesso.")

```
