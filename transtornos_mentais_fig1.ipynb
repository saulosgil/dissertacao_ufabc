# Prevalência global de transtornos mentais
# Figura 1

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Dados
transtornos = [
    "Transtornos alimentares",
    "Esquizofrenia",
    "Transtorno bipolar",
    "Transtorno de conduta",
    "Transtornos do espectro autista",
    "TDAH",
    "Transtorno do desenvolvimento\nintelectual (idiopático)",
    "Outros transtornos mentais",
    "Transtornos depressivos",
    "Transtornos de ansiedade",
]

prevalencias = [0.2, 0.3, 0.5, 0.6, 0.8, 1.1, 1.2, 1.4, 4.0, 4.4]

# Cores: depressão em vermelho-alaranjado vivo, ansiedade em laranja, demais em tons de roxo/azul
cores = [
    "#4B3F8C",   # Transtornos alimentares — azul-marinho escuro
    "#7B6FBE",   # Esquizofrenia — lilás médio
    "#9B8FD0",   # Transtorno bipolar — lilás claro
    "#A8C4D8",   # Transtorno de conduta — azul-cinza claro
    "#5BAFD6",   # TEA — azul-celeste
    "#5C3D9E",   # TDAH — roxo
    "#C086C8",   # Intelectual — lilás rosado
    "#C878B8",   # Outros — rosa-roxo
    "#D63B2F",   # Transtornos depressivos — DESTAQUE vermelho-vivo
    "#E8913A",   # Transtornos de ansiedade — laranja
]

fig, ax = plt.subplots(figsize=(11, 7))
fig.patch.set_facecolor("#FFFFFF")
ax.set_facecolor("#FFFFFF")

y_pos = np.arange(len(transtornos))
bars = ax.barh(y_pos, prevalencias, color=cores, height=0.6, zorder=3)

# Destaque especial para depressão (índice 8)
idx_depressao = 8
bars[idx_depressao].set_edgecolor("#8B0000")
bars[idx_depressao].set_linewidth(2)

# Rótulos de valor no final de cada barra
for i, (bar, val) in enumerate(zip(bars, prevalencias)):
    cor_texto = "#D63B2F" if i == idx_depressao else "#333333"
    peso = "bold" if i == idx_depressao else "normal"
    tamanho = 11 if i == idx_depressao else 9.5
    ax.text(
        val + 0.04, bar.get_y() + bar.get_height() / 2,
        f"{val}%",
        va="center", ha="left",
        fontsize=tamanho, fontweight=peso, color=cor_texto
    )

# Linha de referência vertical suave
for x in [1, 2, 3, 4]:
    ax.axvline(x, color="#DDDDDD", linewidth=0.8, zorder=1)

# Configurações dos eixos
ax.set_yticks(y_pos)
ax.set_yticklabels(transtornos, fontsize=9.5, color="#333333")
ax.set_xlabel("Taxa de prevalência, 2021 (%)", fontsize=10, color="#555555", labelpad=10)
ax.set_xlim(0, 5.1)
ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(["0%", "1%", "2%", "3%", "4%"], fontsize=9, color="#777777")

# Destaque na label do eixo Y para depressão
ytick_labels = ax.get_yticklabels()
for i, label in enumerate(ytick_labels):
    if i == idx_depressao:
        label.set_color("#D63B2F")
        label.set_fontweight("bold")
        label.set_fontsize(10.5)

# Título
ax.set_title(
    "PREVALÊNCIA GLOBAL DE TRANSTORNOS MENTAIS",
    fontsize=13, fontweight="bold", color="#1A1A2E",
    pad=18, loc="center"
)

# Anotação de destaque para depressão
ax.annotate(
    "⚠ Transtornos depressivos:\nCerca de 320 milhões de pessoas",
    xy=(4.0, idx_depressao),
    xytext=(3.2, idx_depressao - 1.6),
    fontsize=8.5,
    color="#D63B2F",
    fontweight="bold",
    arrowprops=dict(arrowstyle="->", color="#D63B2F", lw=1.3),
    bbox=dict(boxstyle="round,pad=0.4", facecolor="#FFF0EE", edgecolor="#D63B2F", linewidth=1)
)

# Remover bordas desnecessárias
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(True)
ax.spines["bottom"].set_visible(True)

ax.tick_params(axis="y", length=0)
ax.tick_params(axis="x", colors="#999999", length=3)

plt.tight_layout(pad=2)
plt.savefig("img/transtornos_mentais.png", dpi=300, bbox_inches="tight")
plt.show()
print("Gráfico salvo com sucesso!")
```
